```python
import yaml
with open('/content/trash_dataset/data.yaml', 'r') as f:
    trash_yaml = yaml.safe_load(f)

trash_yaml['train'] = '/content/trash_dataset/train/images'
trash_yaml['val'] = '/content/trash_dataset/valid/images'
trash_yaml['test'] = '/content/trash_dataset/test/images'
with open('/content/trash_dataset/data.yaml', 'w') as f:
    yaml.dump(trash_yaml, f)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
results = model.train(data='/content/trash_dataset/data.yaml', epochs=20)
```

```python
metrics = model.val(data='/content/trash_dataset/data.yaml', split='test')
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
```

```python
model = YOLO('yolov8n.pt')
results = model.train(data='/content/trash_dataset/data.yaml', epochs=1)
```

```python
img_path = '/content/trash_dataset/test/images/*.jpg'

results = model.predict(img_path)
img = results[0].plot()

import matplotlib.pyplot as plt
plt.imshow(img)
plt.axis('off')
plt.title("YOLOv8 Detection")
plt.show()

import cv2
cv2.imwrite('/content/predictions.png', img)
```

```python
import yaml
plant_yaml = {
    'train': '/content/plant_disease_dataset/train/images',
    'val': '/content/plant_disease_dataset/valid/images',
    'test': '/content/plant_disease_dataset/test/images',
    'nc': 3,
    'names': ['healthy', 'early_blight', 'late_blight']
}

with open('/content/plant_disease_dataset/data.yaml', 'w') as f:
    yaml.dump(plant_yaml, f)
```

```python
seg_model = YOLO('yolov8n-seg.pt')
seg_results = seg_model.train(data='/content/plant_disease_dataset/data.yaml', epochs=20, task='segment')
```

```python
seg_metrics = seg_model.val(data='/content/plant_disease_dataset/data.yaml', split='test')
print("mAP50:", seg_metrics.box.map50)
print("mAP50-95:", seg_metrics.box.map)
```

```python
plant_test_images = glob.glob('/content/plant_disease_dataset/test/images/*.jpg')
img_path = plant_test_images[0]

results = seg_model.predict(img_path, task='segment')
seg_img = results[0].plot()

plt.imshow(seg_img)
plt.axis('off')
plt.title("YOLOv8 Segmentation")
plt.show()
```

```python
cv2.imwrite('/content/predictions_seg.png', seg_img)
```
